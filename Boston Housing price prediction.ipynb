```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn.datasets
from sklearn.model_selection import train_test_split
from xgboost import XGBRegressor
from sklearn import metrics
```

```python
df = pd.read_csv('Boston.csv')
```

```python
print(df)
```

```python
df = df.drop(columns=['Unnamed: 0'], axis = 1, inplace = False)
```

```python
print(df.head())
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
correlation = df.corr()
plt.figure(figsize=(10, 10))
sns.heatmap(correlation, cbar = True, square = True,fmt='.1f',annot=True, annot_kws={'size':8} )
```

```python
X = df.iloc[:,:13]
X
```

```python
Y = df.iloc[:,-1]
Y
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state = 2)
print(X.shape, X_train.shape, X_test.shape)
```

```python
model = XGBRegressor()
model.fit(X_train,Y_train)
```

```python
train_data_prediction = model.predict(X_train)
train_data_prediction
```

```python
score_1 = metrics.r2_score(Y_train, train_data_prediction)
score_2 = metrics.mean_absolute_error(Y_train, train_data_prediction)
print("Rsquared error is : ", score_1)
print("Mean absolute error is : ", score_2)
```

```python
test_data_prediction = model.predict(X_test)
test_data_prediction
score_1 = metrics.r2_score(Y_test, test_data_prediction)
score_2 = metrics.mean_absolute_error(Y_test, test_data_prediction)
print("Rsquared error is : ", score_1)
print("Mean absolute error is : ", score_2)
```

```python
plt.figure(figsize=(10, 6))
sns.regplot(x=Y_test, y=test_data_prediction, scatter_kws={'alpha':0.5})
plt.title('Regression Line Plot')
plt.xlabel('Actual Values')
plt.ylabel('Predicted Values')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.kdeplot(Y_test, label='Actual', shade=True)
sns.kdeplot(test_data_prediction, label='Predicted', shade=True)
plt.title('Distribution of Actual and Predicted Values')
plt.xlabel('Values')
plt.ylabel('Density')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
residuals = Y_test - test_data_prediction
absolute_residuals = np.abs(residuals)
colors = np.where(absolute_residuals <= 2.5, 'blue', 'purple')

plt.scatter(Y_test, residuals, alpha=0.7, c=colors)
plt.axhline(y=0, color='black', linestyle='--', linewidth=2)  # Changed line color to black
plt.title('Residuals Plot')
plt.xlabel('Actual Values')
plt.ylabel('Residuals')
plt.show()
```
